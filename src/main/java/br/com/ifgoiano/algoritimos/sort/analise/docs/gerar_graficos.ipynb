```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

# ── Configuração visual global ──────────────────────────────────────────────
plt.rcParams.update({
    "figure.facecolor":  "#ffffff",
    "axes.facecolor":    "#f8f8f8",
    "axes.edgecolor":    "#cccccc",
    "axes.grid":         True,
    "grid.color":        "#e0e0e0",
    "grid.linestyle":    "--",
    "grid.linewidth":    0.7,
    "font.family":       "DejaVu Sans",
    "font.size":         11,
    "axes.titlesize":    13,
    "axes.titleweight":  "bold",
    "axes.labelsize":    11,
    "xtick.labelsize":   9,
    "ytick.labelsize":   9,
    "legend.fontsize":   10,
    "legend.framealpha": 0.9,
    "lines.linewidth":   2.2,
    "lines.markersize":  5,
})

COR_SEL = "#E24B4A"   # vermelho  — SelectionSort
COR_INS = "#EF9F27"   # âmbar     — InsertionSort
COR_QCK = "#1D9E75"   # verde     — QuickSort

# ── Carregar dados ──────────────────────────────────────────────────────────
df = pd.read_csv("resultados.csv")
df.columns = df.columns.str.strip()

ds1 = df[df["Dataset"] == "Dataset1"].copy()
ds2 = df[df["Dataset"] == "Dataset2"].copy()

col_sel = "SelectionSort(ms)"
col_ins = "InsertionSort(ms)"
col_qck = "QuickSort(ms)"

# ── Helper: formata eixo Y em ms / s ───────────────────────────────────────
def fmt_ms(x, _):
    if x >= 1000:
        return f"{x/1000:.1f}s"
    return f"{x:.0f}ms"

# ── Helper: formata eixo X em k ────────────────────────────────────────────
def fmt_k(x, _):
    if x >= 1000:
        return f"{int(x/1000)}k"
    return str(int(x))

# ════════════════════════════════════════════════════════════════════════════
# 1. Comparação geral (todos os dados, linha do tempo completo)
# ════════════════════════════════════════════════════════════════════════════
def grafico_geral():
    fig, ax = plt.subplots(figsize=(10, 5))

    # Dataset1 + Dataset2 sem repetir o ponto 10k
    d1 = ds1.copy()
    d2 = ds2[ds2["Tamanho"] > 10000].copy()
    todos = pd.concat([d1, d2]).sort_values("Tamanho")

    ax.plot(todos["Tamanho"], todos[col_sel], color=COR_SEL, marker="o", label="SelectionSort")
    ax.plot(todos["Tamanho"], todos[col_ins], color=COR_INS, marker="s", linestyle="--", label="InsertionSort")
    ax.plot(todos["Tamanho"], todos[col_qck], color=COR_QCK, marker="^", label="QuickSort")

    ax.set_title("Comparação geral — todos os tamanhos")
    ax.set_xlabel("Tamanho do vetor")
    ax.set_ylabel("Tempo médio (ms)")
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(fmt_k))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_ms))
    ax.legend()
    fig.tight_layout()
    fig.savefig("grafico_comparacao_geral.png", dpi=150)
    plt.close(fig)
    print("✔ grafico_comparacao_geral.png")

# ════════════════════════════════════════════════════════════════════════════
# 2. Dataset 1 — vetores 1k a 10k
# ════════════════════════════════════════════════════════════════════════════
def grafico_dataset1():
    fig, ax = plt.subplots(figsize=(9, 5))

    ax.plot(ds1["Tamanho"], ds1[col_sel], color=COR_SEL, marker="o", label="SelectionSort")
    ax.plot(ds1["Tamanho"], ds1[col_ins], color=COR_INS, marker="s", linestyle="--", label="InsertionSort")
    ax.plot(ds1["Tamanho"], ds1[col_qck], color=COR_QCK, marker="^", label="QuickSort")

    ax.set_title("Dataset 1 — vetores de 1 000 a 10 000 elementos")
    ax.set_xlabel("Tamanho do vetor")
    ax.set_ylabel("Tempo médio (ms)")
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(fmt_k))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_ms))
    ax.legend()
    fig.tight_layout()
    fig.savefig("grafico_dataset1.png", dpi=150)
    plt.close(fig)
    print("✔ grafico_dataset1.png")

# ════════════════════════════════════════════════════════════════════════════
# 3. Dataset 2 — vetores 10k a 100k
# ════════════════════════════════════════════════════════════════════════════
def grafico_dataset2():
    fig, ax = plt.subplots(figsize=(9, 5))

    ax.plot(ds2["Tamanho"], ds2[col_sel], color=COR_SEL, marker="o", label="SelectionSort")
    ax.plot(ds2["Tamanho"], ds2[col_ins], color=COR_INS, marker="s", linestyle="--", label="InsertionSort")
    ax.plot(ds2["Tamanho"], ds2[col_qck], color=COR_QCK, marker="^", label="QuickSort")

    ax.set_title("Dataset 2 — vetores de 10 000 a 100 000 elementos")
    ax.set_xlabel("Tamanho do vetor")
    ax.set_ylabel("Tempo médio (ms)")
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(fmt_k))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_ms))
    ax.legend()
    fig.tight_layout()
    fig.savefig("grafico_dataset2.png", dpi=150)
    plt.close(fig)
    print("✔ grafico_dataset2.png")

# ════════════════════════════════════════════════════════════════════════════
# 4. Escala logarítmica — todas as curvas legíveis ao mesmo tempo
# ════════════════════════════════════════════════════════════════════════════
def grafico_log():
    d1 = ds1.copy()
    d2 = ds2[ds2["Tamanho"] > 10000].copy()
    todos = pd.concat([d1, d2]).sort_values("Tamanho")

    fig, ax = plt.subplots(figsize=(10, 5))

    ax.plot(todos["Tamanho"], todos[col_sel], color=COR_SEL, marker="o", label="SelectionSort")
    ax.plot(todos["Tamanho"], todos[col_ins], color=COR_INS, marker="s", linestyle="--", label="InsertionSort")
    ax.plot(todos["Tamanho"], todos[col_qck], color=COR_QCK, marker="^", label="QuickSort")

    ax.set_yscale("log")
    ax.set_title("Comparação em escala logarítmica (todos os tamanhos)")
    ax.set_xlabel("Tamanho do vetor")
    ax.set_ylabel("Tempo médio (ms) — escala log")
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(fmt_k))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_ms))
    ax.legend()
    fig.tight_layout()
    fig.savefig("grafico_escala_log.png", dpi=150)
    plt.close(fig)
    print("✔ grafico_escala_log.png")

# ════════════════════════════════════════════════════════════════════════════
# 5. Gráfico de barras — tamanhos selecionados (escala log p/ não esmagar QuickSort)
# ════════════════════════════════════════════════════════════════════════════
def grafico_barras():
    tamanhos_alvo = [1000, 5000, 10000, 50000, 100000]
    todos = pd.concat([ds1, ds2]).drop_duplicates("Tamanho")
    sub = todos[todos["Tamanho"].isin(tamanhos_alvo)].sort_values("Tamanho")

    labels = [fmt_k(t, None) for t in sub["Tamanho"]]
    x = np.arange(len(labels))
    w = 0.25

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.bar(x - w, sub[col_sel], w, label="SelectionSort", color=COR_SEL, alpha=0.85)
    ax.bar(x,     sub[col_ins], w, label="InsertionSort", color=COR_INS, alpha=0.85)
    ax.bar(x + w, sub[col_qck], w, label="QuickSort",     color=COR_QCK, alpha=0.85)

    ax.set_yscale("log")
    ax.set_title("Comparação por tamanho (escala logarítmica)")
    ax.set_xlabel("Tamanho do vetor")
    ax.set_ylabel("Tempo médio (ms) — escala log")
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(fmt_ms))
    ax.legend()
    fig.tight_layout()
    fig.savefig("grafico_barras_comparacao.png", dpi=150)
    plt.close(fig)
    print("✔ grafico_barras_comparacao.png")

# ════════════════════════════════════════════════════════════════════════════
# Executar tudo
# ════════════════════════════════════════════════════════════════════════════
if __name__ == "__main__":
    grafico_geral()
    grafico_dataset1()
    grafico_dataset2()
    grafico_log()
    grafico_barras()
    print("\nPronto! 5 gráficos gerados na pasta atual.")
```
